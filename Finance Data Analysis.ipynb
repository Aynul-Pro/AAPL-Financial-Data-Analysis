Adding percentage change, log return columns with calculating and visualizing moving averages 

```python
data['% change']= data['Close'].pct_change()
```

```python
data.head(10)
```

```python
import numpy as np
```

```python
data['Log returns'] = np.log(data['Close']/data['Close'].shift())
```

```python
data.head(10)
```

```python
data['Log returns'].std()
```

```python
#Visualizing Histogram of the Log Return 
fig, ax = plt.subplots()
data['Log returns'].hist(ax=ax, bins=50, alpha=0.6, color='g')
ax.set_xlabel("Log Return")
ax.set_ylabel("Frequency of log return")
ax.set_title("Histogram of Log Return")
```

```python
# creating 10 day moving average column
data['Ten day MA'] = data['Close'].rolling(10).mean()
```

```python
data.tail()
```

```python
#Creating 10 day Exponential Moving Average
data['Ten day EMA'] = data['Close'].ewm(span=10, adjust=False).mean()
```

```python
data.tail()
```

```python
# Creating Visualization to compare moving averages with daily closing price
fig, ax = plt.subplots()
data[['Ten day MA', 'Ten day EMA','Close']].loc['2020-12-01':].plot(ax=ax)
```
